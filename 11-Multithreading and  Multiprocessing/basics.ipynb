### Program 
Program is a sequence of instruction written in programming languages.

example - Google Chrome  -> .exe file <- Program -> Browser should work.

## Process
A process is simply an instance of an program that is being executed.
* Every process has separate memory space ,
* one process cannot corrupt another process.
* I/O requirements - 
* Increased exceution time to switch between process.
* each process has its own ID.


## Thread
A thread is a unit of execution within a process.
* Single threaded process - This will have its own stack and register , but it will be sharing its code segment , data segment and heap with the process itself.
