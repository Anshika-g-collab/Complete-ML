### Data Visualisation with Seaborn
Seaborn is a Python visualisation library based on MatplotLib that provides a high-level interface for drawig attarctive and informative statistical graphics. Seaborn helps in creating complex visualisations with just a few lines of code.

```python
import seaborn as sns
```

```python
### Basic Plotting with seaborn
tips = sns.load_dataset('tips')
tips
```

```python
## Create a scatter plot
import matplotlib.pyplot as plt
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("scatter Plot of Total_Bill vs Tips")
```

```python
## Line plot 
sns.lineplot(x='size' ,y = 'total_bill',data=tips )
plt.title("Line Plot of Total bill by size")
plt.show()
```

```python
## Categorical plots
## Bar Plot
sns.barplot(x = 'day',y = 'total_bill' , data = tips,color = 'teal')
plt.title('Bar Plot of total bill by day')
plt.show()
```

```python
## Boxplot
sns.boxplot(x = "day",y='total_bill',data = tips,color = 'crimson')
plt.show()
```

```python
## Violin plot
sns.violinplot(x = "day",y='total_bill',data = tips,color = 'yellowgreen')
```

```python
### Histogram 
## kde = kernel density
sns.histplot(tips['total_bill'],bins = 10,kde = True)
```

```python
sns.histplot(tips['total_bill'],bins = 10,kde = False)
```

```python
### KedPlot

sns.kdeplot(tips['total_bill'],fill = True,color = 'crimson')
```

```python
## Pair Plot =  relationship of all the variables with different different variables.
sns.pairplot(tips)
```

```python
## Heatmap

corr = tips[['total_bill','tip','size']].corr()
print(corr)

sns.heatmap(corr,annot=True)
```

```python
import pandas as pd
pulse_df = pd.read_csv('data.csv')
pulse_df.head()
```

```python
## Plot total pulse by calories
plt.figure(figsize=(10,6))
sns.barplot(x = 'Pulse',y='Calories',data = pulse_df,estimator = sum)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x = 'Pulse',y='Duration',data = pulse_df,estimator = sum)
plt.title("Total Pulse by Duration")
plt.xlabel('Pulse')
plt.ylabel('Duration')
plt.show()
```
